## Vectorisation


```python
#Vectorisation avec tf-idf

from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer( max_df=0.6, min_df=0.008) #On a environ 1000 mots uniques pour 10000 reviews
tf_idf = vectorizer.fit_transform(neg_reviews)
array=vectorizer.get_feature_names_out()
print(array)
print(tf_idf)
```

```python
#Conversion tf_idf en matrice
matrice_tfidf=pd.DataFrame(tf_idf.toarray(),columns=array)
matrice_tfidf

```

```python
#NMF
from sklearn.decomposition import NMF
nmf_model = NMF(n_components=15, init='random', random_state=0)
W = nmf_model.fit_transform(tf_idf) # matrice documentsxtopics 
H = nmf_model.components_ # matrice topicsxdocuments
print(W)
print(H)
```

```python
matrice_doctopic=pd.DataFrame(W)
matrice_doctopic
```

```python
matrice_topicdoc=pd.DataFrame(H)
matrice_topicdoc
```

```python
#Affichage des topics
def display_topics(model, feature_names, num_top_words,topic_names=None):
    for ix, topic in enumerate(model.components_):
        #print topic, topic number, and top words
        if not topic_names or not topic_names[ix]:
            print("\nTopic ", ix)
        else:
            print("\nTopic: '",topic_names[ix],"'")
        print(", ".join([feature_names[i] \
             for i in topic.argsort()[:-num_top_words - 1:-1]]))
```

```python
#Affichage des mots corr√©l√©s aux diff√©rents topics

nmf_model = NMF(15)
doc_topic = nmf_model.fit_transform(tf_idf)
display_topics(nmf_model, vectorizer.get_feature_names(), 5)
```
